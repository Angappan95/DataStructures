```python
def factorial(n):
    if n==0:
        return 1
    smallOut = factorial(n-1)
    output = n * smallOut
    return output
```

```python
def natural_sum(n):
    if n == 1:
        return 1
    smallOut = natural_sum(n-1)
    output = n + smallOut
    return output
```

```python
def is_sorted(ls):
    l = len(ls)
    
    if l == 0 or l == 1:
        return True
    
    if ls[0] > ls[1]:
        return False
    
    smaller_list = ls[1:]
    is_smaller_list_sorted = is_sorted(smaller_list)
    
    return is_smaller_list_sorted
```

```python
# Replace character in a string recursively
def replaceRecursively(arr, x, y, index=0):
    l = len(arr)
    if l == 0:
        return -1
    
    if index == l:
        return ''.join(arr)
    
    if arr[index] == x:
        arr[index] = y
    
    smallerList = replaceRecursively(arr, x, y, index+1)
    return smallerList
```

```python
def partition(arr, start, end):
    pivot = arr[start]
    pos = 0
    for temp in range(start, end+1):
        if arr[temp] < pivot:
            pos += 1
    arr[start], arr[start+pos] = arr[start+pos], arr[start]
    
    pivot_index = start+pos
    
    i = start
    j = end
    
    while i < j:
        if arr[i] < pivot:
            i += 1
        elif arr[j] >= pivot:
            j -= 1
        else:
            arr[i], arr[j] = arr[j], arr[i]
            i += 1
            j -= 1

    return pivot_index

def quickSort(arr, start, end):
    if start >= end:
        return 
    
    pivotIndex = partition(arr, start, end)
    quickSort(arr, start, pivotIndex-1)
    quickSort(arr, pivotIndex+1, end)
```

```python
def tower_hanoi(n, a, b, c):
    if n==1:
        print(f'Move 1st disc from {a} -> {c}')
        return 
    tower_hanoi(n-1,  a, c, b)
    print(f'Move {n}th disc from {a} -> {c}')
    tower_hanoi(n-1, b, a, c)
```

```python
def power_opt(num, exp):
    if exp == 0:
        return 1
    small_out = power_opt(num, exp//2)
    if exp%2 == 0:
        return small_out * small_out
    else:
        return num * small_out * small_out
```
