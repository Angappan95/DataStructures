```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
```

```python
from time import sleep
import pandas as pd
```

```python
# Get all items with in page and store in dict
def get_profiles_data(driver):
    profiles = dict()
    items = driver.find_elements(By.ID, 'simple_res_bg')
    for item in items:
        x = item.text.split('\n')
        profile_id = x[0]
        profiles[profile_id] = dict()
        profiles[profile_id]['age'] = ''.join(x[1].split()[2:])
        profiles[profile_id]['height'] = ''.join(x[2].split()[2:])
        profiles[profile_id]['koil'] = ''.join(x[3].split()[2:])
        profiles[profile_id]['star'] = ''.join(x[4].split()[2:])
        profiles[profile_id]['education'] = ''.join(x[5].split()[2:])
        profiles[profile_id]['occupation'] = ''.join(x[6].split()[2:])
        profiles[profile_id]['income'] = ''.join(x[7].split()[2:])
        profiles[profile_id]['rasi'] = ''.join(x[8].split()[2:])
        profiles[profile_id]['location'] = ''.join(x[9].split()[2:])
        profiles[profile_id]['last_viewed'] = ''.join(x[10].split()[2:])
    return profiles
```

```python
data = dict()
```

```python
# Collect data for first page
data.update(get_profiles_data(driver))

# Iterate over next 'n' pages
for _ in range(5):
    sleep(30)
    driver.find_element(By.LINK_TEXT, 'Next').click()
    data.update(get_profiles_data(driver))
```

```python
df = pd.DataFrame.from_dict(data).T
df.to_csv('out.csv')
```
